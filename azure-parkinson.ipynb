```python
# summary table function
def summary(df):
    print(f'data shape: {df.shape}')
    summ = pd.DataFrame(df.dtypes, columns=['data type'])
    summ['#missing'] = df.isnull().sum().values * 100
    summ['%missing'] = df.isnull().sum().values / len(df)
    summ['#unique'] = df.nunique().values
    desc = pd.DataFrame(df.describe(include='all').transpose())
    summ['min'] = desc['min'].values
    summ['max'] = desc['max'].values
    summ['first value'] = df.loc[0].values
    summ['second value'] = df.loc[1].values
    summ['third value'] = df.loc[2].values
    
    return summ
```

```python
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python
conditions = [
    (defog_m['StartHesitation'] == 1),
    (defog_m['Turn'] == 1),
    (defog_m['Walking'] == 1)]
choices = ['StartHesitation', 'Turn', 'Walking']
defog_m['event'] = np.select(conditions, choices, default='Normal')
```

```python
defog_m['event'].value_counts().to_frame()
```

```python
train_df = defog_m[['AccV','AccML','AccAP','event']]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

train_df['target'] = le.fit_transform(train_df['event'])
```

```python
X = train_df.drop(['event','target'], axis=1)
y = train_df['target']
```

```python
X
```

```python
y
```

```python
from imblearn.over_sampling import SMOTE

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Check class distribution before applying SMOTE
print("Class distribution before SMOTE:")
print(y_train.value_counts())

# Find the size of the majority class (Class 0)
majority_class_size = y_train.value_counts().max()

# Apply SMOTE to the training data with sampling_strategy set to the desired ratio
smote = SMOTE(sampling_strategy={1: majority_class_size, 2: majority_class_size, 3: majority_class_size})
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Check class distribution after applying SMOTE
print("\nClass distribution after SMOTE:")
print(pd.Series(y_train_resampled).value_counts())
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier

# Initialize the Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the classifier
rf_classifier.fit(X_train_resampled, y_train_resampled)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report  # Import the accuracy_score function
```

```python
y_pred = rf_classifier.predict(X_test)
```

```python
def evaluate_metrics(y_true, y_pred):
    
    # Calculate metrics
    accuracy = accuracy_score(y_true, y_pred)
    conf_matrix = confusion_matrix(y_true, y_pred)
    classification_rep = classification_report(y_true, y_pred)

    # Print the metrics
    print(f'Accuracy: {accuracy}')
    print(f'Confusion Matrix:\n{conf_matrix}')
    print(f'Classification Report:\n{classification_rep}')
```
